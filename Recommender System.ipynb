```python
movierating_matrix = movies_data.pivot_table(values = 'Rating', index = 'User_ID', columns = 'Movie_Title')
```

```python
movierating_matrix
```

```python
movie_ratings.sort_values('Rating_Count', ascending = False).head(20)
```

```python
pulp_fiction_rating = movierating_matrix['Pulp Fiction (1994)']
pulp_fiction_rating.head(10)
```

```python
pulp_movie_corr = movierating_matrix.corrwith(pulp_fiction_rating)
```

```python
corr_df = pd.DataFrame(pulp_movie_corr, columns = ['Correlation'])
```

```python
corr_df.head()
```

```python
corr_df = corr_df.join(movie_ratings['Rating_Count'])
```

```python
corr_df.head()
```

```python
corr_df[corr_df['Rating_Count']>500].sort_values(['Correlation'], ascending = False)
```

## Prediction Function

```python
def predict_movie(movie_name):
    movie_rating = movierating_matrix[movie_name]
    movie_corr = movierating_matrix.corrwith(movie_rating)
    movie_corr_df = pd.DataFrame(movie_corr, columns = ['Correlation'])
    movie_corr_df = movie_corr_df.join(movie_ratings['Rating_Count'])
    predictions = movie_corr_df[movie_corr_df['Rating_Count']>500].sort_values(['Correlation'], ascending = False)
    
    return predictions
```

```python
predictions = predict_movie("Man of the House (1995)") 
```
